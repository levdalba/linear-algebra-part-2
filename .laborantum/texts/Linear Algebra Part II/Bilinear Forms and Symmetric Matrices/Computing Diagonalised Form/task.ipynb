```python
def diagonalize(A):
    ### YOUR CODE HERE ###
    
    Lambda, Q = np.linalg.eigh(A)

    idx = np.argsort(Lambda)[::-1]
    Lambda = Lambda[idx]
    Q = Q[:, idx]

    for i in range(Q.shape[1]):
        if Q[0, i] < 0:
            Q[:, i] *= -1

    L, D, _ = ldl(A)

    def check_definiteness(values):
        if np.all(values > 0):
            return '+'
        elif np.all(values < 0):
            return '-'
        elif np.all(values >= 0):
            return '0+'
        elif np.all(values <= 0):
            return '0-'
        else:
            return '+-'

    definiteness = check_definiteness(Q)

    return Lambda, Q, D, L, definiteness
```

```python
inputs = json_tricks.load('inputs.json')

outputs = []

for input in inputs:
    outputs.append(diagonalize(**input))

json_tricks.dump(outputs, '.answer.json')
```
