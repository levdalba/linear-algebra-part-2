```python
def SVD(A):
    ### YOUR CODE HERE ###
    ATA = A.T @ A
    AAT = A @ A.T

    U, S, Vt = np.linalg.svd(A, full_matrices=False)

    idx = np.argsort(S)[::-1]
    S = S[idx]
    U = U[:, idx]
    Vt = Vt[idx, :]

    for i in range(Vt.shape[0]):
        if Vt[i, 0] < 0:
            Vt[i, :] *= -1

    for i in range(min(U.shape[1], len(S))):
        if U[i, 0] < 0:
            U[:, i] *= -1

    return ATA, AAT, U, np.diag(S), Vt
```

```python
inputs = json_tricks.load('inputs.json')
outputs = []

for input in inputs:
    U1, S1, V1 = np.linalg.svd(input['A'], full_matrices=True)
    res = SVD(**input)
    outputs.append(res)

    ## CHECK THAT U @ S @ V.T is equal to A

    U = res[2]
    S = res[3]
    V = res[4]

    print('Reconstruction misfit: {:.2}'.format((input['A'] - U @ S @ V).mean()))

json_tricks.dump(outputs, '.answer.json')
```
