```python
import numpy as np
import json_tricks
```

```python
def one_step(A):
    Q, R = np.linalg.qr(A)
    B = np.dot(R, Q)
    return B

def compute_dif(neig, leig, B):
    dif = 0
    for i in range(len(B)):
        dif += abs(neig[i] - leig[i])
    return dif

def compute_dev(B):
    cur_dev = 0
    for i in range(len(B)):
        for j in range(i):
            cur_dev = max(cur_dev, abs( B[i, j] ))

    return cur_dev

def qr_iteration(A):
    A = A.astype('float64')

    devs = []
    
    cur=compute_dev(A)
    devs.append(cur)
    
    B=one_step(A)
    leig = np.diag(B)

    cur=compute_dev(B)
    devs.append(cur)
    dif = 1

    while dif > (1e-10):
        B=one_step(B)

        neig = np.diag(B)

        dif = compute_dif(neig, leig, B)

        leig = neig

        cur_dev = compute_dev(B)

        devs.append(cur_dev)

    print(devs)

    return leig, devs
```

```python
from tqdm import tqdm

inputs = json_tricks.load('inputs.json')

results = []
for args in tqdm(inputs):
    res = qr_iteration(**args)
    results.append(res)

json_tricks.dump(results, '.answer.json')
```
